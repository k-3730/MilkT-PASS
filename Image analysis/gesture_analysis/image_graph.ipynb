### 제스처 분석 시각화

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.ensemble import IsolationForest
from settings import subject, load_path, save_path, PATH



save_path = f'{save_path}'
sub_list = subject
motion_dict = {"english" : [], 'math' : [], "korean" : []}

for sub in sub_list:

    for lecture in os.listdir(f"{save_path}\\{sub}"):
        df = pd.read_csv(f"{save_path}\\{sub}\\{lecture}")
        df = df.drop(columns=['time', 'Unnamed: 0'])
        df = df.dropna()
        df = df[df['motion magnitude'] > 100]


        # 선택한 특성만 있는 데이터프레임 생성
        X = df['motion magnitude'].values.reshape(-1, 1)

        # Isolation Forest 모델 생성
        model = IsolationForest()

        # 모델 훈련 및 예측
        model.fit(X)
        df['is_outlier'] = model.predict(X)  # 예측 결과를 새로운 열로 추가

        # 이상치가 아닌 데이터만 선택
        normal_data = df[df['is_outlier'] == 1]

        # 필요에 따라 이상치 데이터 분석
        outlier_data = df[df['is_outlier'] == -1]
        count = len(outlier_data)
        mean = df['motion magnitude'].mean()


        print(lecture)
        print(f"모션 평균값: {mean}")
        # print(f'표쥰편차 : {std_dev}')
        # print(f"강조 모션 개수: {count}")
        print(round((count/len(df))*100, 2))
        motion_dict[sub].append(round((count/len(df))*100, 2))
```
